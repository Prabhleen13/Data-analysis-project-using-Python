```python
avg_usage_by_gender = df.groupby('Gender')['Avg_Daily_Usage_Hours'].mean()

plt.figure(figsize=(6, 4))
avg_usage_by_gender.plot(kind='bar', color=['pink', 'yellow'])
plt.title("Bar Chart: Average Daily Usage by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Daily Usage (Hours)")
plt.show()
```

```python
platform_counts = df['Most_Used_Platform'].value_counts()
top4 = platform_counts[:4]
others = platform_counts[4:].sum()

platform_data = top4.copy()
platform_data['Others'] = others
plt.figure(figsize=(6,5))
plt.pie(platform_data, labels=platform_data.index,autopct='%1.1f%%')
plt.title("Most Used Social Media Platforms")
plt.show()
```

```python
plt.figure(figsize=(7,4))
mean_data=df.groupby('Sleep_Hours_Per_Night')['Mental_Health_Score'].mean()
plt.plot(mean_data.index, mean_data.values, color='purple', linestyle='-')
plt.title("Sleep Hours vs Mental Health Score")
plt.xlabel("Sleep Hours Per Night")
plt.ylabel("Mental Health Score")
plt.show()
```
